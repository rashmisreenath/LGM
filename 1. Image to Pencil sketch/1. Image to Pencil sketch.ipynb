### Beginner Level Task 4: Image to Pencil Sketch with Python

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import scipy.ndimage
import cv2 
from skimage import io
from PIL import Image
```

### Reading the image

```python
image=cv2.imread("D:/Down/tiger.jpeg")
image
```

```python
plt.imshow(image)
```

### Converting the image to gray scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

### Inverting the grey scale image

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image)
```

### Blurred image using gaussian function

```python
blurred = cv2.GaussianBlur(inverted_image, (25, 25), 0)
plt.imshow(blurred)
```

### Inverting the blurred image

```python
inverted_blurred = 255 - blurred
```

### Dividing the gray image by inverted blurred image to get the final output

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.imshow(pencil_sketch)
```

```python
cv2.imshow("original image", image)
#cv2.imshow("New Tiget", gray_image)
#cv2.imshow("Inverted", inverted_image)
#cv2.imshow("blurred", blurred)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```
